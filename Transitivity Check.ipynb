```python
import itertools

def generate_dice(n, k): #same as in program
    # Generate all possible combinations of n integers that sum to k
    combinations = list(itertools.combinations(range(1, k), n - 1)) #generates all combos of integers for n(sides) (except for 0)
    
    dice_combos = [] #empty list that stores dice combos that work
    for combo in combinations:#loop
        die = [combo[0]] #starts list for die
        for i in range(1, n - 1): #iterates the for loop using the range specified in parenthesis
            die.append(combo[i] - combo[i - 1]) #finds number of pips on each side (takes difference between elements to ensure # of pips >=1)
        die.append(k - combo[-1]) #finds pips on last side by using total number of pips(k) - last element of the combination
        
        # Check if every side has at least 1 pip
        if all(side >= 1 for side in die): #makes sure all die have at least one side
            die.sort()
            if die not in dice_combos:
                dice_combos.append(die) #adds die to dice list
    
    return dice_combos
    

def compare_die(d1, d2): #same as in program
    face1wins = 0
    face2wins = 0
    tie = 0

    for face1 in d1:
        for face2 in d2:
            if face1 > face2:
                face1wins += 1
            if face2 > face1:          
                face2wins +=1
            if(face1 == face2):
                tie +=1
    if face1wins > face2wins:
        return 1
    if face2wins > face1wins:
        return 2
    if face1wins == face2wins:
        return 0

     

def isitTransitive(n, k): #prints the 3 compared dice that directly violate rules of transivity
    all_dice = generate_dice(n, k)  #generate all dice for n and k

    for dice1 in all_dice:
        for dice2 in all_dice: #we are first going to check if the conditions of transitivity apply
            if compare_die(dice1, dice2) == 1:  # check if dice1 beats dice2
                for dice3 in all_dice:
                    if compare_die(dice2, dice3) == 1:  # check if dice2 beats dice3
                        if compare_die(dice1, dice3) != 1: # now check for a violation of transivity (dice1 loses to or ties with dice3)
                            print("Dice1:", dice1, ", Dice 2:", dice2, ", Dice 3:", dice3) #print violating cases 

# Call the function to check transitivity violations
isitTransitive(3,9)
```
